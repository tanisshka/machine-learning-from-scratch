```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
df=pd.DataFrame(iris['data'],columns=['sepal length','sepal width','petal length','petal width'])
df['Target']=iris['target']
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
#Train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split( X,y,test_size=0.25,random_state=42)
```

```python
#Train the model
from sklearn.tree import DecisionTreeClassifier
treeClassifier=DecisionTreeClassifier()
treeClassifier.fit(X_train,y_train)
```

```python
#Visualize the tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

```python
#Predict the data
y_pred=treeClassifier.predict(X_test)
```

```python
#Performance metrics
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)

score=accuracy_score(y_test,y_pred)
print("Model accuracy: ",score*100)

print(classification_report(y_test,y_pred))
```

## Tree Prunning

### Pre-Prunning

```python
pre_prun_classifier=DecisionTreeClassifier(max_depth=2)
pre_prun_classifier.fit(X_train,y_train)
```

```python
#Visualize the model
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(pre_prun_classifier,filled=True)
```

```python
#Predict 
pre_prun_ypred=pre_prun_classifier.predict(X_test)
```

```python
#Performance metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
cm=confusion_matrix(y_test,pre_prun_ypred)
print(cm)

score=accuracy_score(y_test,pre_prun_ypred)
print("Model accuracy: ",score*100)

print(classification_report(y_test,pre_prun_ypred))
```

### Post - Prunning and Hyperparameter Tunning

```python
params={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['sqrt', 'log2'],
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
treeModel=DecisionTreeClassifier()
grid=GridSearchCV(treeModel,param_grid=params,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
hyp_y_pred=grid.predict(X_test)
```

```python
#Performance metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
cm=confusion_matrix(y_test,hyp_y_pred)
print(cm)

score=accuracy_score(y_test,hyp_y_pred)
print("Model accuracy: ",score*100)

print(classification_report(y_test,hyp_y_pred))
```
