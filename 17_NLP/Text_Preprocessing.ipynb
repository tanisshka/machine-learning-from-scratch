## Text Preprocessing Stemming

### Stemming
Stemming is the process of reducing words to their root words by removing suffix and prefix

Example:
1. playing -> play
2. played -> play
3. plays -> play
4. studies -> studi

This improves:
1. Search engines
2. Chatbots
3. Text classification
4. NLP models

Most common stemming algorithms
1. Porter Stemmer (most popular)
2. Snowball Stemmer
3. Lancaster Stemmer

### Porter Stemming

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
porter_stemming=PorterStemmer()
for word in words:
    print(word+" ------> "+porter_stemming.stem(word))
```

### RegexpStemmer

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")
```

### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
snowball_stemming=SnowballStemmer(language='english')
```

```python
for word in words:
    print(word+"------>"+snowball_stemming.stem(word))
```

```python
print(snowball_stemming.stem("fairly"))
```

```python
print(porter_stemming.stem("fairly"))
```

## Text Preprocssing Lemmatization
Lemmatization is the process of converting a word into its base (dictionary) form, called a lemma, while considering the word’s meaning and part of speech (POS).

It is smarter than just cutting words — it understands grammar.

Lemmatization = Reducing a word to its meaningful root form using vocabulary and grammar rules.

| Original Word | Lemmatized Form |
| ------------- | --------------- |
| running       | run             |
| better        | good            |
| studies       | study           |
| was           | be              |
| mice          | mouse           |


```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

## Text Preprocessing Stopwards
Stopwords are commonly used words in a language that usually don’t carry significant meaning in text analysis.

Examples in English:

a, an, the, is, am, are, was, were, in, on, at, of, for, to, and, but

🔹 Why Are They Removed?

Because in many NLP tasks they:
1. Appear very frequently
2. Add noise
3. Increase vocabulary size
4. Do not help in distinguishing documents

Example:

Sentence:

"The movie was very good and the acting was amazing"

After removing stopwords:

"movie good acting amazing"

Now it's more meaningful for ML.



```python
para='''
This is a story about a young student who was in a small college and who was trying to learn about artificial intelligence and machine learning. He was very interested in the field, but he was also a little confused about how to start and where to begin. In the beginning, he was reading a lot of articles and watching many videos, and he was trying to understand what the experts were saying. However, he did not always understand everything that they were talking about, and so he was feeling a bit lost at times. But he knew that if he kept on working hard and if he continued to practice every day, he would eventually improve and become better at what he was doing.
'''
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
from nltk.tokenize import sent_tokenize
sentences=sent_tokenize(para)
```

```python
for sentence in sentences:
    print(sentence+"\n")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

for i in range(len(sentences)):
    words = word_tokenize(sentences[i].lower())
    
    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words and word not in string.punctuation
    ]
    
    sentences[i] = " ".join(words)
    print(sentences[i])
```

## Named Entity Recognition
Named Entity Recognition (NER) is an NLP task that identifies and classifies important entities in text into predefined categories.

📌 Example

Sentence:

import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

NER Output:
| Word       | Entity Type        |
| ---------- | ------------------ |
| Elon Musk  | PERSON             |
| SpaceX     | ORG (Organization) |
| California | GPE (Location)     |
| 2002       | DATE               |

So instead of just treating words as tokens, NER understands what they represent.

Real-world applications:
1. Resume parsing (extract name, skills, college)
2. News analysis (who did what, where)
3. Financial analysis (company names, amounts)
4. Legal document processing
5. Chatbots
6. Healthcare systems

If you build a resume skill extractor → you are using NER.

```python
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```
